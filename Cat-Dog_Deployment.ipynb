```python
%%writefile app.py
import streamlit as st
import tensorflow as tf
import numpy as np
from PIL import Image, ImageOps
import cv2 as cv
class_name = ["Cat", "Dog"]

@st.cache_resource()

def load_model():
  model = tf.keras.models.load_model('/content/drive/MyDrive/cat_dog_cnn/cat_dog_99.h5')
  return model

with st.spinner("Model is Being Loaded......"):
  model = load_model()


st.write("""
# Cat-Dog Classification
""")


file = st.file_uploader('Please Upload an Image file', type=["jpg","jpeg","png"])

def import_and_predict(image_data,model):
  size = (224,224)
  image = ImageOps.fit(image_data, size, Image.ANTIALIAS)
  image = np.asarray(image)
  img = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  img_reshape = img[np.newaxis,...]
  prediction = model.predict(img_reshape)
  return prediction

if file is None:
  st.text("Please Provide FIle First..!!!")
else:
  image = Image.open(file)
  st.image(image,width = 400)
  predictions = import_and_predict(image, model)
  score = tf.nn.sigmoid(predictions[0])
  idx = np.argmax(score)
  cls = class_name[idx]
  st.write(f"# This image is Most Likely {cls} With an Accuracy of {np.round(np.max(score)*100,2)}%")
```
